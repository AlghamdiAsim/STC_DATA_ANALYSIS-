```python
from sklearn.metrics.pairwise import cosine_similarity

# Create User-Item Matrix
user_item = df.pivot_table(index='user_id_maped',
                           columns='program_name',
                           values='rating',
                           fill_value=0)

# Calculate Item Similarity
item_similarity = pd.DataFrame(
    cosine_similarity(user_item.T),
    index=user_item.columns,
    columns=user_item.columns
)

# Function to recommend programs for a user
def recommend_user(user_id, top_n=5):
    if user_id not in user_item.index:
        return "User not found."

    watched = user_item.loc[user_id]
    watched = watched[watched > 0].index

    recommendations = pd.Series(dtype=float)

    for program in watched:
        recommendations = pd.concat([recommendations, item_similarity[program]])

    recommendations = recommendations.groupby(recommendations.index).mean()
    recommendations = recommendations.drop(watched, errors='ignore')

    return recommendations.sort_values(ascending=False).head(top_n)

# for Example
print("Recommended programs for user 26138:")
print(recommend_user(26138))
```

```python
# Top 5 recommendations for users who watched "Moana"

top5_recommendations = item_similarity["Moana"] \
    .sort_values(ascending=False) \
    .drop("Moana") \
    .head(5)

print("Top 5 recommendations for people who watched 'Moana':")
print(top5_recommendations)
```
